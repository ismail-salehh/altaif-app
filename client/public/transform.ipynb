```python
import os
from PIL import Image
```

```python
# Folder containing your images
input_folder = "images"       # Change to your folder name
output_folder = "converted2"   # Output folder name

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Target size
TARGET_SIZE = (600, 600)

# Supported formats
valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif', '.webp')

for filename in os.listdir(input_folder):
    if filename.lower().endswith(valid_extensions):
        img_path = os.path.join(input_folder, filename)
        if filename=='1-sibling.jpg' or filename=='loud-noise.jpg':
            continue
        img = Image.open(img_path)

        # Resize only if either dimension is greater than 600
        if img.width > 600 or img.height > 600:
            img.thumbnail(TARGET_SIZE, Image.Resampling.LANCZOS)

        # Convert and save as .webp
        base_name = os.path.splitext(filename)[0]
        output_path = os.path.join(output_folder, f"{base_name}.png")

        img.save(output_path, "PNG", quality=90)
        print(f"✅ Converted: {filename} → {output_path}")

print("\nAll images processed successfully!")
```
